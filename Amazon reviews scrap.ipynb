# Scrapping reviews from flipkart -PRODUCT--Iphone 13

```python
list_review=[] #initializing list of reviews
#Starting for loop through all the pages to scrape reviews from all 954 pages
for page in range(2,954):
    SAMPLE_URL1 ="https://www.flipkart.com/apple-iphone-13-midnight-128-gb/product-reviews/itmca361aab1c5b0?pid=MOBG6VF5Q82T3XRS&lid=LSTMOBG6VF5Q82T3XRSOXJLM9&marketplace=FLIPKART&page="+str(page)
    r = requests.get(SAMPLE_URL1)    #getting request from flipkart
    soup = BeautifulSoup(r.content, 'html.parser') #Extracting content through beautiful soup
    co=1 #counter of no of reviews Scraped
    for i in (soup.find_all(class_='t-ZTKy')): #Specific tag class of flipkart page where each review is stored              
        co+=1 #incrementing review counter
        list_review.append(i.text) #Appending reviews to list.
```

```python
#creating dictionary to load all the reviews in data frame
dic = {'review':list_review,
      'length':[len(i) for i in list_review]}
```

```python
frame=pd.DataFrame(dic) #through pandas creating dataframe
```

# TEXT mining

```python
reviews_data = pd.read_csv('REVIEWS DATA.csv') #raeding file
```

```python
reviews_data = reviews_data.drop(['Unnamed: 0'],axis=1) #droping unneccesary column
```

```python
#normalizing all the tweets
import string
p = string.punctuation
remv_punc = str.maketrans("", "", p)


from nltk.stem import PorterStemmer
ps = PorterStemmer()

#TEXT PROCESSING---------------
#REMOVING PUNCHUATIONS
#CONVERTING INTO LOWER LETTERS
#REMOVING WHITESPACES
for i in range(len(reviews_data)):
    reviews_data.review[i] = reviews_data.review[i].lower()
    reviews_data.review[i] = reviews_data.review[i].translate(remv_punc)
    reviews_data.review[i] = reviews_data.review[i].replace("  ", " ").strip()
    
    
```

```python
reviews_data
```

```python
reviews = (list(reviews_data.review)) #LISTING ALL REVIEWS
```

```python
from nltk import word_tokenize
```

```python
#Save the NOUNS only
review_noun=[]
for i in reviews:
    token=word_tokenize(i) #spliting words or we can say tokenizing
    tag=nltk.pos_tag(token) #pos taging
    for j in range(len(tag)):
        if tag[j][1]=='NN': #extracting nouns only
            review_noun.append(tag[j][0])
    
```

```python
unique=list(set(review_noun))    
#creating unique lsit
```

```python
len(unique)
```

```python
#craeting dictionary of data of each noun in which review it has occured binarily
dic={}
for nou in unique:
    li=[]
    for i in reviews:
        
        if nou in word_tokenize(i.lower()):
            li.append(1)
        else:
            li.append(0)
        dic[nou]=li    
```

```python
data = pd.DataFrame(dic)
```

```python
data
```

```python
data.to_csv('dataframe_noun_without_stemming.csv')
```

# Compactness Pruning 

```python
pair_nouns = apr_df.copy() #creating copy
```

```python
pair_nouns.itemsets[0]
```

```python
pair = []
for i in range(len(pair_nouns.itemsets)):
    if len(pair_nouns.itemsets[i]) == 2:
        pair += [pair_nouns.itemsets[i]]
```

```python
pair[0]
```

```python
for i in range(len(pair)):
    pair[i] = list(pair[i])
```

```python
pair[0]
```

```python
compact = []
for i in range(len(pair)):
        for sentence in reviews:
            sent = word_tokenize(sentence)
            if pair[i][0] in sent and pair[i][1] in sent:
                count = sent.index(pair[i][0]) - sent.index(pair[i][1])
                if abs(count) <4:
                    compact.append((pair[i],sentence))
```

```python
len(compact)
```

```python
all_features = []
for i in compact:
    all_features.append(i[0])
```

```python
features =[]
feature_dic = {}
for i in all_features:
    if i not in features:
        features.append(i)
        feature_dic[str(i[0])+' '+str(i[1])] = 1
    else :
        feature_dic[str(i[0])+' '+str(i[1])] = feature_dic.get(str(i[0])+' '+str(i[1]))+1
```

```python
features
```
